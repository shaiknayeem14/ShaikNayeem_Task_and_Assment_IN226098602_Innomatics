```python
def check_stock(items):
    for product, qty in items.items():
        if qty<15:
            print(product,"-> Reorder Alert:")
        else:
            print(product,"-> Stock ok")
inventory={
    "Rice":10,
    "Wheat":25,
    "Sugar":12
}
check_stock(inventory)
```

```python
def result(marks):
    total=0
    for m in marks:
        total +=m
    avg=total/len(marks)
    if avg >=50:
        print("Pass")
    else:
        print("Fail")
result([60,40,50])
```

```python
def fare(km, peak):
    total=50+(12*km)
    if peak:
        total=total*1.25
    return total
while True:
    km=int(input("Enter distance(km):"))
    peak=input("Peak hour? (yes/no)")
    if peak=="yes":
        p=True
    else:
        p=False
    print("Fare = ", fare(km,p))
    again=input("Retry? (yes/no):")
    if again=="no":
        break
```

```python
def check(att):
    present=0
    for i in att:
        if i=="P":
            present +=1
    #checking eligibility criteria
    percent=(present/len(att))*100
    if percent >=75:
        return "Eligible"
    else:
        return "Not Eligible"
#takie input from the user 
att = input("Enter attendance (P/A separated by space): ").split()
print(check(att))
```

```python
def check(pw):
    #if password length is less than 8 it should return weak
    if len(pw)<8:
        return "Weak"
    digit=False
    special=False
    #letting the password has digits and special char for the strong password
    for ch in pw:
        if ch.isdigit():
            digit=True
        if ch in "@#$":
            special=True
    #checkinh wheather the password is strong or weak
    if digit and special:
        return "Strong"
    else:
        return "Weak"
pw = input("Enter password: ")
print(check(pw))
```
