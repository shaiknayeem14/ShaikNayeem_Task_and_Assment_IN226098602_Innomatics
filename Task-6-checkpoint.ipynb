```python
#Problem 1: Smart Parking Lot Management System
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    status = "Available" if parked <= capacity else "Full"

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)
smart_parking(50, ["IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#Problem 2: Online Food Delivery Time Estimator
def delivery_eta(distance, traffic, weather):
    eta = distance * 5   # base time
    if traffic == "High":
        eta += 20
    elif traffic == "Medium":
        eta += 10
    if weather == "Rainy":
        eta += 15
    elif weather == "Storm":
        eta += 20
    print("Estimated Delivery Time:", eta, "minutes")
# Function Call
delivery_eta(7, "High", "Rainy")
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"
    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)
# Function Call
seat_occupancy(300, [1] * 150)
```

```python
#Problem 4: Cloud Server Load Classification System
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg:.0f}%")
    print("Server Status:", status)


# Function Call
server_load([45, 60, 70, 85, 90])
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor
def resource_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Function Call
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
#Problem 6: Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Function Call
event_registration(90, 105)
```
