```python
#Problem 1: Unique Words in a Sentence
sentence_input = input("Enter sentence: ")
# Convert everything to lowercase for uniform comparison
cleaned_sentence = sentence_input.lower()
word_list = cleaned_sentence.split()
unique_word_set = set(word_list)
print("Unique words count:", len(unique_word_set))
print("Unique words:", unique_word_set)
```

```python
#Problem 2: Highest Salary from Employee Data
employee_data = {
    "hari": 63000,
    "krishna": 34000,
    "harsha": 52000
}
highest_salary = 0
highest_employee = ""
for name, salary in employee_data.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [14,50,25,51]
# Initialize max and min with first element
maximum_value = numbers[0]
minimum_value = numbers[0]
# Loop through the list starting from second element
for num in numbers[1:]:
    if num > maximum_value:
        maximum_value = num
    if num < minimum_value:
        minimum_value = num
# Display results
print("List:", numbers)
print("Max:", maximum_value)
print("Min:", minimum_value)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [500, 1400, 999, 1300, 200]
threshold = 1000
count = 0
for price in prices:
    if price > threshold:
        count += 1
print("Products above", threshold, ":", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]
# Count total days
total_days = len(attendance)
# Count number of presents
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1
# Calculate percentage
attendance_percentage = (present_count / total_days) * 100
# Display result
print("Attendance Percentage:", attendance_percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Display result
print("Unique phone numbers:", unique_numbers)
```

```python
#Problem 7: Count Character Frequency

text = input("Enter a string: ")

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print("Character Frequency:", char_count)
```

```python
#Problem 8: Convert List to Tuple
numbers_list = [30,40,50]
print("Original List:", numbers_list)
numbers_tuple = tuple(numbers_list)
print("Converted Tuple:", numbers_tuple)
print("Data type after conversion:", type(numbers_tuple))
```

```python
#Problem 9: Check if a Key Exists in a Dictionary
employees = {
    "hari": 63000,
    "krishna": 34000,
    "harsha": 52000
}

search_key = "harsha"

if search_key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
